### Chapter 02. 크리스마스의 기온 변화를 그래프로 그리기

```python
# split 연습
print("1990-09-01".split('-'))
print("1990-09-01".split('-')[0])
print("1990-09-01".split('-')[1])
print("1990-09-01".split('-')[2])
```

```python
import csv

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean_temp = [] # 평균 기온 데이터를 저장할 리스트
max_temp = [] # 최고 기온 데이터를 저장할 리스트
min_temp = [] # 최저 기온 데이터를 저장할 리스트


for row in data:
    if row[4] != '': # 최고 기온 데이터가 존재한다면
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25':
            max_temp.append(float(row[4])) # result 리스트에 최고 기온 값을 추가
            mean_temp.append(float(row[2])) # result 리스트에 최고 기온 값을 추가
            min_temp.append(float(row[3])) # result 리스트에 최고 기온 값을 추가
    
f.close()
#print(len(result))
#print(result)
```

```python
plt.rcParams['font.family'] = 'AppleGothic' # 한글 폰트를 쓰기 위함
plt.figure(figsize=(10,4)) # 단위: inch
plt.title('크리스마스 기온 변화')

plt.plot(mean_temp, label = 'mean') # x 좌표 값 생략 가능
plt.plot(max_temp, label = 'max')
plt.plot(min_temp, label = 'min')

plt.xlabel("Day(12/25)")
plt.ylabel("Temperature")

plt.legend()
plt.grid()
plt.show()
```

> 연평균 기온 변화 그래프 그려보기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

year = [] # 꺾은 선 그래프를 그릴 떼, x축 좌푯값으로 활용할 예정!
day = [] # 365일치의 기온 값을 저장할 때 활용
mean = [] # 꺾은 선 그래프를 그릴 때, y축 좌푯값으로 활용할 예정

for row in data:
    y = int(row[0].split('-')[0]) # str -> int (연도 정보를 추출함)
    if y not in [1907, 1950, 1951, 1952, 1953, 2024]:
        if row[2] != '':
            day.append(float(row[2])) # 기온이기에 타입 변환함
            if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '31':
                mean.append(sum(day) / len(day))  # y 좌푯값으로 이용
                year.append(y) # x 좌푯값으로 이용
                # 연도가 갱신될 때 리스트 초기화해야 함!
                day = []
                
    elif y == 2024:
        break
                
f.close()
```

```python
plt.figure(figsize=(10,4))
plt.title('연평균 기온 변화 그래프')
plt.plot(year, mean, marker = '.')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.show()
```
