```python
import random

dice = []
for i in range(0, 10, 1):
    dice.append(random.randint(1,6))

# 경험적 확률: 시행에 따라서 계산되는 확률
# 수학적 확률: 1/6
# 큰 수의 법칙: 

plt.figure()
plt.hist(dice, bins=6)
plt.grid(True)
plt.xlabel('Dice')
plt.ylabel('Count')
plt.show()
```

```python
import csv
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

august= []
january= []

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '08':
            august.append(float(row[4])) # str -> float  
        elif row[0].split('-')[1] == '01':
            january.append(float(row[4]))

f.close()

plt.figure()

plt.hist(august, bins=100, color='red', label='August') # 8월의 히스토그램 
plt.hist(january, bins=100, color='blue', label='January') # 1월의 히스토그램
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)
       # 1월,2월....
month = [[],[],[],[],[],[],[],[],[],[],[],[]] # 12개의 빈 리스트 저장

for row in data:
    if row[4] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[4])) # 1을 왜 빼???? 나 이해 못했는데?????????

f.close()

plt.figure(figsize=[10,3])
plt.boxplot(month, showfliers=False) # 이상치 제거
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python

#plt.style.use('ggplot') # R 프로그래밍에서 사용
plt.style.use('default')

plt.figure(figsize=[10,3])
plt.boxplot(month, showfliers=False) # 이상치 제거
plt.xlabel('Temperature')
plt.ylabel('Day')
plt.grid(True)
plt.show()
```
