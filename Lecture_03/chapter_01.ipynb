### Chapter 01. 기온 공공데이터

```python
import csv
# 스트림 개방
df = open("seoul.csv", "r", encoding="cp949") # r: read mode, w: write mode
data = csv.reader(df)

print(data)

# 조건문을 통해 결측값 확인하기
for row in data:
    if '' in row:
        print(row)
    
# 스트림 폐쇄
df.close()
```

```python
import pandas as pd

df = pd.read_csv("seoul.csv", encoding="cp949")
print(df.head())  # 상위 5개 행 출력
```

```python
print(df.isnull().sum())

# 결측값이 포함된 행만 출력
print(df[df.isnull().any(axis=1)])
```
