### Chapter 02. 인구 구조를 다양한 형태로 시각화하기

#### 1. 막대 그래프 (Bar Graph)

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar([0,1,3,5,8,9], [1,2,3,4,5,6]) # 두 개의 전달 인자값을 필요로 한다.
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조를 알고싶은 지역명을 입력하세요.: ')
result = []
for row in data:
    if name in row[0]:
        for j in row[3:]:
            result.append(int(j)) # str -> int
f.close()


# 시각화 함수 사용하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.figure(figsize=(2,5))
plt.barh(range(0,101,1), result)
plt.title(name + '의 인구수 파악해보기')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.grid(True)
plt.show()
```

#### 2. 성별로 나눠 파악해보기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조를 알고싶은 지역명을 입력하세요.: ')
male = []
female = []

for row in data:
    if name in row[0]:
        for j in row[3:104]: # 남자 인구수
            male.append(-int(j))  # - 를 붙여 왼쪽으로 그래프가 그려지게 함
            
        for j in row[106:207]: # 여자 인구수
            female.append(int(j))
            
f.close()


# 시각화 함수 사용하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.barh(range(0,101,1), male, label='남자')
plt.barh(range(0,101,1), female, label='여자')
plt.title(name + '의 남녀별 인구수 파악해보기')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.grid(True)
plt.legend()
plt.show()
```
