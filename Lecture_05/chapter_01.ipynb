### Chapter 01. 우리 동네 인구 조사하기

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조를 알고싶은 지역명을 입력하세요.: ')
result = []
for row in data:
    if name in row[0]:
        for j in row[3:]:
            result.append(int(j)) # str -> int
f.close()


# 시각화 함수 사용하기

plt.rcParams['font.family'] = 'AppleGothic'
plt.figure()
plt.plot(result)
plt.title('인구수 파악해보기')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.grid(True)
plt.show()
```
