> 데이터 병합하기

- **merge**

```python
# 1번 테이블
s1 = pd.read_excel('Dataset/nation.xlsx')
s1
```

```python
# 2번 테이블
s2 = pd.read_excel("Dataset/code.xlsx")
s2
```

```python
# inner join으로 병합
pd.merge(s1, s2, how="inner", on="국적코드")
```

- **grouping**

```python
s1.groupby("국적코드").sum()
```

```python
s1.groupby(['국적코드','성별']).sum()
```

     대분류를 국적코드, 중분류를 성별로 설정

```python
s1.groupby("국적코드")[["입국객수", "증가수"]].mean()
```

     ⚠️ 문자형 열이 섞여있어 mean() 함수를 썼을 때 에러가 난듯 하다. mean 함수로 확인할 숫자형 열만 지정해줬다.
