### Chapter 01. 우리 동네 인구 조사하기

> 우리 동네 인구 구조 시각화하기

    풍납2동의 인구구조를 파악하기로 결정

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)
name = input('인구 구조를 알고싶은 지역명을 입력하세요.')
result = []

for row in data:
    if row[0] == name:
        for j in row[3:]:
            result.append(int(j)) # str --> int
f.close()

# 데이터 시각화하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.figure()
plt.title('인구 수 파악하기')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.grid(True)
plt.plot(result)
plt.show()
    
```

    '==' 을 사용할 경우 주소가 행에 적혀있는 것과 정확히 동일해야 데이터가 출력됨. 따라서 in 연산자를 사용해서 다시 도전

```python
print('풍납동' in '서울특별시 송파구 풍납2동')
print('송파구' in '서울특별시 송파구 풍납2동')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('age.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)
name = input('인구 구조를 알고싶은 지역명을 입력하세요.')
result = []

for row in data:
    if name in row[0]:
        for j in row[3:]:
            result.append(int(j))

f.close()

# 데이터 시각화하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.figure()
plt.plot(result)
plt.title(name + '의 연령별 인구 수 파악하기')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.grid(True)
plt.show()
```
