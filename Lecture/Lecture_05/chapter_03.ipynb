### Chapter 03. 우리 동네 인구 구조를 파이 차트로 나타내기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조를 알고싶은 지역명을 입력하세요.: ')
male = []
female = []

for row in data:
    if name in row[0]:
        for j in row[3:104]: # 남성의 나이가 저장된 열 범위
            male.append(-int(j))
            
        for j in row[106:207]: # 여성의 나이가 저장된 열 범위
            female.append(int(j))
f.close()

# 데이터 시각화하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.title(name + '의 남녀 인구 수 파악하기')
plt.barh(range(0, len(male), 1), male, label='남성', color='skyblue')
plt.barh(range(0, len(female), 1), female, label='여성', color='red')
plt.xlabel('인구 수')
plt.ylabel('나이')
plt.grid(True)
plt.legend()
plt.show()
    

```

#### ✅ replace 함수
    replace(A, B)는 문자열의 A를 모두 B로 변경한다.

```python
'1,865'.replace(',', '---')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조를 알고싶은 지역명을 입력하세요.: ')
male = []
female = []

for row in data:
    if name in row[0]:
        for j in row[3:104]: # 남성의 나이가 저장된 열 범위
            j = j.replace(',', '')
            male.append(-int(j))
            
        for j in row[106:207]: # 여성의 나이가 저장된 열 범위
            j = j.replace(',', '')
            female.append(int(j))        
        break
        
f.close()

# 데이터 시각화하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.title(name + '의 남녀 인구 수 파악하기')
plt.barh(range(0, len(male), 1), male, label='남성', color='skyblue')
plt.barh(range(0, len(female), 1), female, label='여성', color='red')
plt.xlabel('인구 수')
plt.ylabel('나이')
plt.grid(True)
plt.legend()
plt.show()
```

#### ✅ 파이 차트 (Pie Chart)
    - 원 그래프
    - 3시 방향부터 순차적으로 비율을 나타낸다. 방향은 반시계 방향으로 간다.

```python
plt.figure()

data = [4, 7, 11, 25]
labels = ['A형', 'B형', 'O형', 'AB형']
color = ['skyblue', 'purple', 'limegreen', 'yellow']

# explode: 폭발시키다, 원하는 레이블을 강조할 때 주로 사용
plt.pie(data, labels=labels, autopct='%.1f%%', colors=color, explode=(0.3,0,0,0))

plt.legend(loc=10) # Location: n.위치 , 반시계 방향으로 돌아감
plt.show()
```

> 제주도의 성별 인구 비율 표현하기

    예로부터 제주도는 돌, 바람, 여성이 많다하여 삼다도라고 불렀다고 한다. 실제로 여성의 인구수가 더 많은지 분석해보자.

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('인구 구조를 알고싶은 지역명을 입력하세요.: ')
# 파이차트는 비율을 나타내기 위해 사용하므로 더 이상 리스트의 형태가 아님
male = 0
female = 0

for row in data:
    if name in row[0]:
        for j in row[3:104]: # 남성의 나이가 저장된 열 범위
            j = j.replace(',', '')
            male += int(j)
            
        for j in row[106:207]: # 여성의 나이가 저장된 열 범위
            j = j.replace(',', '')
            female += int(j)        
        break
        
f.close()

print('남성의 총 인구수:', male)
print('여성의 총 인구수:', female)
```

```python
# 데이터 시각화하기
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

plt.figure()
plt.title(name + '의 남녀 인구 수 파악하기')

plt.pie([male, female], 
        labels=['남성', '여성'], 
        colors=['blue', 'red'], 
        autopct='%.2f%%',
        startangle=90) # 남성의 비율이 3시 방향부터 시작하니 90도만큼 위치를 옮겨준다.
plt.legend()
plt.show()
```

    남성과 여성의 성비가 거의 동일함
    🚩 요 근래는 삼다도라는 표현은 더 이상 적절하지 않은 것 같다.
