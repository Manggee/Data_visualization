### **Chapter 02. 데이터 확인하기**

```python
import pandas as pd

dust = pd.read_excel("dust.xlsx")
dust.tail(3)
```

```python
dust.info()
```

```python
dust.describe()
```

```python
print(dust.columns)
```

```python
dust.rename(columns={'날짜':"date", 
                     '아황산가스':"so2", 
                     '일산화탄소':"co", 
                     '오존':"o3", 
                     '이산화질소':"no2"}, inplace=True)
dust.head(3)
```

```python
dust.info()
```

```python
dust["date"] = dust["date"].str[0:10]   # year-month-day 만 추출!

dust.head(3) 
```

```python
dust.info()
```

```python
dust["date"] = pd.to_datetime(dust["date"])
dust.info()   # object(str) --> datetime
```

```python
print(dust["date"].dt.year)
```

```python
print(dust["date"].dt.month)
```

```python
print(dust["date"].dt.day)
```

```python
dust.head(3)
```

```python
dust["year"] = dust["date"].dt.year
dust["month"] = dust["date"].dt.month
dust["day"] = dust["date"].dt.day
dust.head(3)
```

```python
print(dust.columns)
```

```python
dust = dust[['date', 'year', 'month', 'day', 'so2', 'co', 'o3', 'no2', 'PM10', 'PM2.5']]

dust.head(3)
```

```python
# Not Available --> NA --> 결측값
dust.isna().sum()
```

```python
dust.head(5)
# NaN --> Not a Number --> 결측값
```

```python
# dust.dropna()

# forward + fill --> "ffill"
# backward fill --> "bfill"

dust = dust.ffill()
dust.head(5)
```

```python
dust = dust.bfill()  # backward fill
dust.head(5)
```

```python
# (X) dust.fillna(method="pad")
# (O) dust.ffill()
# (O) dust.bfill()

dust.isna().sum()
```

```python
weather = pd.read_excel("weather.xlsx")
weather
```

```python
weather.info()
```

```python
weather.drop(["지점", "지점명"], axis=1, inplace=True)
weather.head(3)
```

```python
weather.columns = ["date", "temp", "wind", "rain", "humid"]
weather.head(3)
```

```python
weather.info()
```

```python
weather["date"] = weather["date"].dt.date
weather.head(3)
```

```python
weather.info()
```

```python
weather["date"] = pd.to_datetime(weather["date"])
weather.info()
```
